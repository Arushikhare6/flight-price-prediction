# Notebook 03: Data Cleaning

## Objective

In this notebook, we will:

- Handle missing values
- Remove duplicate records
- Drop unnecessary columns
- Verify the cleaned dataset
- Save the cleaned dataset for the next phase

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
DATA_DIR = Path("../data/raw")

df = pd.read_excel(DATA_DIR / "Train.xlsx")

df.head()
```

```python
print("Missing Values:\n")
df.isnull().sum()
```

```python
df.dropna(inplace=True)

print("Dataset Shape:", df.shape)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)

print("Dataset Shape After Removing Duplicates:", df.shape)
```

```python
df["Additional_Info"].value_counts()
```

```python
df.drop(columns=["Additional_Info"], inplace=True)

df.head()
```

```python
print("Final Shape:", df.shape)

print("\nMissing Values:\n")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
PROCESSED_DIR = Path("../data/processed")
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

df.to_csv(PROCESSED_DIR / "cleaned_train.csv", index=False)

print("Cleaned dataset saved successfully!")
```

# Data Cleaning Summary

### Completed Tasks

- Removed missing values
- Removed duplicate records
- Dropped the `Additional_Info` column
- Saved the cleaned dataset

The dataset is now ready for **Feature Engineering**, where we will extract useful numerical features from dates, times, and durations.
